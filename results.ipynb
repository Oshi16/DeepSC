```python
def split(string):
    string = string[1:-1]
    return [float(value.strip()) for value in string.split(' ') if value.strip()]

split('''[0.41071999 0.62963774 0.7059885  0.83427356 0.87361763 0.87616841
 0.90326034]''')
```
